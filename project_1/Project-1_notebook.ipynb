Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def workExperience(workExpString:str) -> int:
    """ Функция находит опыт работы в месяцах по шаблону строки

    Args:
        workExpString (str): Строка, содержащая опыт работы

    Returns:
        int: Опыт работы (месяц)
    """
    
    # Проверка на пропуск в данных
    if (workExpString is np.nan) or (workExpString == 'Не указано'):
        return np.nan
    
    # Выделяем количество лет и месяцев с учетом специальных условий
    expList = workExpString.split()[:6] # Список, в котором будет нужный нам опыт работы
    
    months = 0 
    years = 0
    if 'месяц' in expList[-1]:
        months = int(expList[-2])
        years = int(expList[2])
    elif 'месяц' in expList[3]:
        months = int(expList[2])
    else:
        years = int(expList[2])
    
    return years*12 + months


# Выделяем опыт работы
resume_data['Опыт работы (месяц)'] = resume_data['Опыт работы'].apply(workExperience)

# Удаляем лишнюю информацию
resume_data = resume_data.drop('Опыт работы', axis=1)  
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def city(cityString:str) -> str:
    """Функция возвращает название города, если он является Москвой или Санкр-Петербургом
    или относится к городам-миллионникам, иначе возвращает название города, как "другие"

    Args:
        cityString (str): Строка с названием города
        
    Returns:
        str: Город или его категория
    """
    # Список городов-миллионников
    million_cities = [
        'Новосибирск', 'Екатеринбург', 'Нижний Новгород',
        'Казань', 'Челябинск', 'Омск',
        'Самара', 'Ростов-на-Дону', 'Уфа',
        'Красноярск', 'Пермь', 'Воронеж', 'Волгоград'
    ]
    
    city = cityString.split(' , ')[0] # Название города
    if city in ['Москва', 'Санкт-Петербург']:
        return city
    
    if city in million_cities:
        return 'город-миллионник'
    
    return 'другие'


def removal(cityString:str) -> bool:
    """Функция возвращает готовность работника к переездам

    Args:
        cityString (str): Строка, содержащая информацию о готовности к переезду

    Returns:
        bool: True - работник готов к переезду, False - в противном случае
    """
    # Сохраняем информацию о переезде, в зависимости от указания станции метро
    if 'м.' in cityString.split(' , ')[1]:
        readyToMove = cityString.split(' , ')[2] # Строка с информацией о готовности к переезду
    else:
        readyToMove = cityString.split(' , ')[1]
    
    return False if readyToMove.split()[0] == 'не' else True


def businessTrip(cityString:str) -> bool:
    """Функция возвращает готовность работника к командировкам

    Args:
        cityString (str): Строка, содержащая информацию о готовности к командировкам

    Returns:
        bool: True - работник готов к командировкам, False - в противном случае
    """
    readyTripList = cityString.split(' , ')[-1].split() # Список слов из строки о командировках
    
    # Проверка, есть ли информация о командировках
    if 'командировкам' not in readyTripList:
        return False
    
    return False if readyTripList[0] == 'не' else True
    
    
# Выделяем город
resume_data['Город'] = resume_data['Город, переезд, командировки'].apply(city)
resume_data['Город'] = resume_data['Город'].astype('category')

# Выделяем готовность к переезду
resume_data['Готовность к переезду'] = resume_data['Город, переезд, командировки'].apply(removal)

# Выделяем готовность к командировкам
resume_data['Готовность к командировкам'] = resume_data['Город, переезд, командировки'].apply(businessTrip)

# Удаляем лишнюю информацию
resume_data = resume_data.drop('Город, переезд, командировки', axis=1)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<center><img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg></center>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Список вариантов занятости
employmentLst = [
    'полная занятость', 'частичная занятость', 'проектная работа',
    'волонтерство', 'стажировка'
]

# Список вариантов графика
scheduleLst = [
    'полный день', 'сменный график',
    'гибкий график', 'удаленная работа',
    'вахтовый метод'
]

# One Hot Encoding преоразования
for employment in employmentLst:
    resume_data[employment] = resume_data['Занятость'].apply(
        lambda x: True if employment in x.split(', ') else False
    )
    
for schedule in scheduleLst:
    resume_data[schedule] = resume_data['График'].apply(
        lambda x: True if schedule in x.split(', ') else False
    )
    
# Удаляем лишнюю информацию
resume_data = resume_data.drop(['График', 'Занятость'], axis=1)
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Импортируем таблицу с курсами валют и удаляем лишнее
exchangeRates = pd.read_csv(
    r'data\ExchangeRates.csv'
)
exchangeRates = exchangeRates.drop(['per', 'time', 'vol'], axis=1)

# Преобразуем дату
resume_data['Обновление резюме'] = pd.to_datetime(
    resume_data['Обновление резюме'],
    dayfirst=True
).dt.date

resume_data['Обновление резюме'] = pd.to_datetime(
    resume_data['Обновление резюме']
)

exchangeRates['date'] = pd.to_datetime(
    exchangeRates['date'],
    dayfirst=True
)

# Выделяем указанную валюту и ЗП
resume_data['Валюта'] = resume_data['ЗП'].apply(
    lambda x: x.split()[-1]
)

resume_data['ЗП'] = resume_data['ЗП'].apply(
    lambda x: int(x.split()[0])
).astype('int32')

# Словарь перевода валюты к ISO формату
ISODict = {curr: curr for curr in dict(resume_data['Валюта'].value_counts()).keys()}
ISODict['бел.руб.'] = 'BYN'
ISODict['грн.'] = 'UAH'
ISODict['сум'] = 'UZS'

resume_data['Валюта'] = resume_data['Валюта'].apply(
    lambda x: ISODict[x]
)

# Объединение таблиц по дате и валюте
resume_data = pd.merge(
    resume_data,
    exchangeRates,
    how='left',
    left_on=['Обновление резюме', 'Валюта'],
    right_on=['date', 'currency']
)

# Указываем курс и пропорцию для рубля
resume_data.loc[resume_data['Валюта'] == 'руб.', ['close', 'proportion']] = 1

# Вычисляем ЗП в рублях
resume_data['ЗП (руб)'] = resume_data['ЗП'] * resume_data['close'] / resume_data['proportion']

# Удаляем лишнее
resume_data = resume_data.drop(
    ['ЗП', 'Валюта', 'currency', 'date', 'close', 'proportion'],
    axis=1
)
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# Логарифмируем шкалу возраста
log_age = np.log(resume_data['Возраст'])

# Строим график
fig = px.histogram(
    log_age,
    nbins=100
)

fig.add_vline(x=log_age.mean(), line_color='red')
fig.add_vline(x=log_age.mean() - 3*log_age.std(), line_color='red')
fig.add_vline(x=log_age.mean() + 3*log_age.std(), line_color='red')

fig.update_layout(
    title='Распределение возраста (логарифмический масштаб)',
    yaxis_title='Количество соискателей',
    xaxis_title='Возраст',
    width=1000,
    height=800
)

fig.show()
# fig.write_html('plotly/LogAgeDistribution.html')

# Асимметрия 
# log_age.skew()
# 0.463

# Определение выбросов
def outliers_z_score(data, feature,
                     log_scale=False, left=3, right=3):
    """Функция находит выбросы методом z-отклонений

    Args:
        data (pd.DataFrame): Исходный DataFrame
        feature (str, optional): Признак для очистки
        log_scale (bool): Требуется ли логарифмирование признака. Defaults to False.
        left (int, optional): Левая граница распределения. Defaults to 3.
        right (int, optional): Правая граница распределения. Defaults to 3.

    Returns:
        pd.DataFrame: Очищенный DataFrame и сами потенциальные выбросы
    """
    if log_scale:
        x = np.log(data[feature])
    else:
        x = data[feature]
        
    mu = np.mean(x) # Находим среднее и стандартное отклонение
    sigma = np.std(x)
    
    # Определяем верхнюю и нижнюю границы распределения
    upper_bound = mu + right*sigma 
    lower_bound = mu - left*sigma
    
    # Отсеиваем аномальные данные
    outliers = data[(x<lower_bound) | (x>upper_bound)]
    cleaned = data[(x>lower_bound) & (x<upper_bound)]
    
    return outliers, cleaned


outliers, resume_data = outliers_z_score(resume_data, 'Возраст', log_scale=True, right=4)

print('Соискатели, данные о которых считаются аномальными по возрасту:')
display(outliers)
```
